# Finding Association for Store Dataset
- Support of `1.5%`
- Confidence of `40%`
- For $T(X)$ be number of transactions containing X, its also the support of $X$
$$
    \implies T(X)=Support(X)
$$
## Support
$$
    Support(X\rightarrow Y) = \frac{Support(X\bigcup Y)}{Support(All)}
$$
## Confidence
$$
    Confidence(X\rightarrow Y) = \frac{Support(X\bigcup Y)}{Support(X)}
$$
## Lift
$$
    Lift(X\rightarrow Y) = \frac{Confidence(X\rightarrow Y)}{Support(X\rightarrow Y)}
$$

# Class struct of Node for FP Tree

```python
class Node:
    def __init__(self,name):
        self.name=name
        self.childNodes={}
        self.count=1
    
    def __hash__(self):
        return hash(str(self.name))
    
    def __eq__(self,other):
        if type(self)==type(other):
            return self.name==other.name
        return self.name==other
    
    def getName(self):
        return str(self.name)
    
    def setParent(self,parent):
        self.parentNode=parent
    
    def _addChild(self,childNode):
        self.childNodes[childNode]=childNode
        childNode.setParent(self)
        if self.parentNode!=self:
            self.count+=1
        return childNode
        
    def _removeChile(self,childNode):
        if self.has(childNode):
            self.childNodes.remove(childNode)
    
    def has(self,childNode):
        if childNode in self.childNodes:
            return True
        return False
    
    def traverseToChild(self,childNode):
        if self.has(childNode):
            self.count+=1
            return (self.childNodes[childNode],True)
        childNode = Node(childNode)
        self._addChild(childNode)
        return (self.childNodes[childNode],False)
```
